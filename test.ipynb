```python
import torch
import numpy as np
from benchmarks import find_and_process_file
from env import Env
from kim2016 import Kim2016
from lin2015 import Lin2015
import pandas as pd

# Example usage
folder_path = "./Lee_instances"  # Replace with the folder containing your files
inst_type = "random"

results = {}

n_rows = 16
for n_tiers in [6,8]:
    for n_bays in [1,2,4,6,8,10]:
        for id in range(1,6):
            if n_tiers == 8 and n_bays in [8,10]:
                continue
            
            container_tensor, file_name = find_and_process_file(folder_path, inst_type, n_bays, n_rows, n_tiers, id)
            lin = Lin2015(pr=30, pb=300)
            cost, moves = lin.run(container_tensor, n_bays, n_rows)
            
            results[file_name] = {
                'cost':cost,
                'moves':moves
            }

df = pd.DataFrame.from_dict(results, orient='index')
df.to_excel('./tmp.xlsx', index=True)
```
